```python
import pandas as pd
from drawings import DrawingPlots
from sklearn.metrics import f1_score
```

```python
def draw_plots(path2json, path2imgs):
    df = pd.read_json(path2json)
    dp = DrawingPlots(df)
    all_plots = dp.scatter()
    for i in range(len(all_plots)):
        all_plots[i].savefig(f'{path2imgs}\\{i}.png')
    return path2imgs
```

```python
path2json = 'deviation.json'
path2folder = 'plots'
draw_plots(path2json, path2folder)
```

```python
def mse(predicts, targets):
    """
    Функция для вычисления Mean Squared Error (MSE).
    :param predicts: список с прогнозами модели
    :param targets: список с правильными ответами
    :return: число - MSE
    """
    n = len(predicts)
    if n != len(targets):
        raise ValueError("Длины двух списков должны быть одинаковы")
    
    return sum([(p - t) ** 2 for p, t in zip(predicts, targets)]) / n
```

```python
df = pd.read_json(path2json)
print(f"mse loss: {mse(df['rb_corners'], df['gt_corners'])}")
print(f"f1 score: {f1_score(df['gt_corners'], df['rb_corners'], average='weighted')}")
```
